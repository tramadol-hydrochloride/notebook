```
import os
from PIL import Image
from google.colab import drive, files

drive.mount('/content/drive')
```

```
img_path = '/content/drive/My Drive/gan_log_01/video10000_x128/preview/image00025000.png'
file_root, _ = os.path.splitext(os.path.basename(img_path))
save_dir = '/content/drive/My Drive/gan_log_01/video10000_x128/preview/{}'.format(file_root)
os.makedirs(save_dir, exist_ok=True)

grid_img = Image.open(img_path)
grid_size = 1280
img_size = 128

imgs = grid2img(grid_img, grid_size, img_size)

for i, img in enumerate(imgs):
    img.save('{}/{}_{}.jpg'.format(save_dir, file_root, str(i).zfill(4)))
```

```
def grid2img(grid_img, grid_size, img_size):
    ret_imgs = []
    for w in range(0, grid_size, img_size):
        for h in range(0, grid_size, img_size):
            ret = Image.new('RGB', (img_size, img_size))
            ret.paste(grid_img, (-(w), -(h)))
            ret_imgs.append(ret)
    return ret_imgs
```
