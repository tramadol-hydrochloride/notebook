```
import os
import glob
import zipfile
from PIL import Image
from tqdm import tqdm
from google.colab import drive

drive.mount('/content/drive')
```

```
# write zip file
dst_dir = '/content/drive/My Drive/sngan/datasets'
data_dir = '/content/drive/My Drive/gan_dataset/video10000_x128'
imgs = glob.glob('{}/*.jpg'.format(data_dir))

with zipfile.ZipFile('{}/video10000_x128.zip'.format(dst_dir), 'w', compression=zipfile.ZIP_DEFLATED) as new_zip:
    for img in tqdm(imgs):
        new_zip.write(img, arcname=os.path.basename(img))
```

```
# read zip file
with zipfile.ZipFile('{}/video10000_x128.zip'.format(dst_dir)) as existing_zip:
    existing_zip.extractall('{}/video10000_x128'.format(dst_dir))
```
