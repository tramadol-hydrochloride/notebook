```
'''
make a text file that contains a image path and its label for the data processing of the conditional dataset for gan training

img01.jpg 0
img02.jpg 0
img03.jpg 1
...
img99.jpg 99
'''
```

```
import os
import shutil
import numpy as np
from google.colab import drive

drive.mount('/content/drive')
files = os.listdir('/content/drive/My Drive/gan_dataset/video10000_x256')
```

```
# if the class (category) has less than 15 images, these images will be categorized as 'others'
# 'merge_list' contains the list of those classes

dirname_to_file_num = {}
merge_list = []

for f in files:
    dirname = '_'.join(f.split('_')[:-1])
    if dirname not in dirname_to_file_num:
        dirname_to_file_num[dirname] = 1
    elif dirname in dirname_to_file_num:
        dirname_to_file_num[dirname] += 1

for d, n in dirname_to_file_num.items():
    if n < 15:
        merge_list.append(d)
```

```
dirname_to_label = {}
label = 0
for f in files:
    dirname = '_'.join(f.split('_')[:-1])
    if dirname in merge_list: 
        dirname = 'others'
    if dirname not in dirname_to_label:
        dirname_to_label[dirname] = label
        label += 1 
```

```
image_list = []
for f in files:
    dirname = '_'.join(f.split('_')[:-1])
    if dirname in merge_list:
        dirname = 'others'
    label = dirname_to_label[dirname]
    image_list.append([f, label])

image_list = np.array(image_list, np.str)
np.savetxt('image_list_video10k.txt', image_list, fmt="%s")
```
