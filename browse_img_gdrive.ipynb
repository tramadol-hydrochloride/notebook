```
import os
import shutil
import numpy as np
from glob import glob
from PIL import Image
from tqdm import tqdm
from google.colab import drive

drive.mount('/content/drive')
```

```
# returns the grid of the images which contain the queried text in the queried directory
def grid_images_query(dirname, query):
    path_list = [path for path in glob(dirname + '/*.jpg') if query in path]
    ret = Image.new('RGB', (1000, 1000), (0, 0, 0))
    idx = 0
    for w in tqdm(range(0, 1000, 100)):
        for h in range(0, 1000, 100):
            if idx > len(path_list) - 1:
                return ret
            im = Image.open(path_list[idx])
            im = im.resize((100, 100))
            ret.paste(im, (w, h))
            idx += 1
    return ret

# returns the grid of the images in the given Google Drive directory
def grid_images(dirname):
    ret = Image.new('RGB', (1000, 1000), (0, 0, 0))
    path_list = glob('{}/*jpg'.format(dirname))
    idx = 0
    for w in tqdm(range(0, 1000, 100)):
        for h in range(0, 1000, 100):
            if idx > len(path_list) - 1:
                return ret
            im = Image.open(path_list[idx])
            im = im.resize((100, 100))
            ret.paste(im, (w, h))
            idx += 1
    return ret
```
